```python
import json
import pandas as pd
```

```python
data_terror = json.load(open("/home/tim/Documents/terrorismus.json"))
data_extrem = json.load(open("/home/tim/Documents/extremism.json"))
data_islam = json.load(open("/home/tim/Documents/Islamism.json"))
```

```python
data_terror.keys()
```

```python
df_terror = pd.DataFrame(data_terror["items"])
df_extrem = pd.DataFrame(data_extrem["items"])
df_islam = pd.DataFrame(data_islam["items"])
```

```python
df_terror
```

```python
df_terror_edit = df_terror[["label", "author", "year", "type"]] #neuer DF mit weniger Spalten
```

```python
df_terror_edit
```

```python
#Bereinigung des Datensatzes
df_terror_edit = df_terror_edit.loc[df_terror_edit["author"].isnull() == False] #alle NaN in Spalte "author" entfernen
df_terror_edit.loc[df_terror_edit.type == "Publication"] #in der Spalte "type" nur "Publication" anzeigen
df_terror_edit = df_terror_edit.drop_duplicates(subset = "label", keep = "first") #Entfernen von Duplikaten
df_terror_edit
```

```python
#Hinzufügen einer neuen Spalte mit Suchbegriff "Terrorismus"
df_terror_edit["search_term"] = "Terrorismus"
df_terror_edit
```

```python
#Umbenennen der Spalten zur besseren Lesbarkeit über ein Dict
df_terror_edit = df_terror_edit.rename(columns={"label": "Title", "author": "Authors", "year": "Year", "type": "Type", "search_term": "Search Term"})
df_terror_edit 
```

```python
#Selber Ablauf für die anderen beiden DF

#df_extrem
df_extrem_edit = df_extrem[["label", "author", "year", "type"]]
df_extrem_edit = df_extrem_edit.loc[df_extrem_edit["author"].isnull() == False] #alle NaN in Spalte "author" entfernen
df_extrem_edit.loc[df_extrem_edit.type == "Publication"] #in der Spalte "type" nur "Publication" anzeigen
df_extrem_edit = df_extrem_edit.drop_duplicates(subset = "label", keep = "first") #Entfernen von Duplikaten
df_extrem_edit["search_term"] = "Extremism" #Neue Spalte mit Suchbegriff
df_extrem_edit = df_extrem_edit.rename(columns={"label": "Title", "author": "Authors", "year": "Year", "type": "Type", "search_term": "Search Term"}) #Umbenennen der Spalten

#df_islam

df_islam_edit = df_islam[["label", "author", "year", "type"]]
df_islam_edit = df_islam_edit.loc[df_islam_edit["author"].isnull() == False] #alle NaN in Spalte "author" entfernen
df_islam_edit.loc[df_islam_edit.type == "Publication"] #in der Spalte "type" nur "Publication" anzeigen
df_islam_edit = df_islam_edit.drop_duplicates(subset = "label", keep = "first") #Entfernen von Duplikaten
df_islam_edit["search_term"] = "Islamism" #Neue Spalte mit Suchbegriff
df_islam_edit = df_islam_edit.rename(columns={"label": "Title", "author": "Authors", "year": "Year", "type": "Type", "search_term": "Search Term"}) #Umbenennen der Spalten


df_islam_edit
```

```python
df_extrem_edit
```

```python
#Zusammenführen der DFs

df_complete_1 = pd.concat([df_extrem_edit, df_islam_edit, df_terror_edit])
df_complete_1
```

```python
#Bereinigen von "falschen" Werten
# 1. Year

df_complete_1["Year_correct"] = df_complete_1.apply(lambda row: row.Year.isdigit(), axis=1)
df_complete_1.loc[df_complete_1.Year_correct == False]
df_complete_1 = df_complete_1.loc[df_complete_1.Year_correct == True]

df_complete_1
```

```python
#Entfernen von Werten in "Year" die nicht mit dem Format YYYY übereinstimmen, also mehr als 4 Zeichen haben

df_complete_1 = df_complete_1[df_complete_1["Year"].str.len() <= 4]
df_complete_1
```

```python
#Entfernen von Werten ind "Authors" die nur ein Wort beinhaltet als nicht aus Vor- und Nachname bestehen
def has_full_name(author_list):
    return any(len(name.split()) > 1 for name in author_list)

df_complete_1 = df_complete_1[df_complete_1["Authors"].apply(has_full_name)]
df_complete_1
    
```

```python
#Die Zählung bei 1 Beginnen lassen
df_complete_1.reset_index(drop=True, inplace=True)
df_complete_1.index = df_complete_1.index + 1
df_complete_1

```

```python

#Aufsteigend nach Jahr sortierten
df_complete_1 = df_complete_1.sort_values(by="Year")
df_complete_1
```

```python
#Spalte "Year_correct" entfernen

df_complete_1 = df_complete_1.drop(columns=["Year_correct"])
df_complete_1
```

```python
#Zuletzt löschen von allen möglichen Zeilen in denen möglicherweise NaN vorkommt (es gab keine)

df_complete_1 = df_complete_1.dropna()
df_complete_1
```

```python
#DF als JSON exportieren
df_complete_1.to_json("Bibsonomy_List.json", orient="records") #Was bedeutet orient?
```

```python
#Überprüfen ob es geklappt hat
with open ("Bibsonomy_List.json", "r") as f:
    j_load = json.load(f)
    print(json.dumps(j_load, indent=2))
```
