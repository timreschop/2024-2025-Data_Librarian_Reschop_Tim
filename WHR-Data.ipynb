# Anpassen des DataFrames auf eigene Bedürfnisse

## Filtern auf das Jahr 2024 (aktuellste Werte) und Ausgabe

```python
df_whr_2024 = df_whr[df_whr["Year"] == 2024]
df_whr_2024
```

## Umbenennen von Spalten

```python
df_whr_2024.rename(columns={"Ladder score" : "Happiness Score"}, inplace=True)
df_whr_2024
```

## Löschen nicht benötigter Spalten

```python
df_whr_2024.drop(columns=["upperwhisker", "lowerwhisker"], inplace=True)
df_whr_2024
```

# Berechnen statistischer Grundwerte

## Mittelwert des Happiness Score (weltweit)

```python
mittelwert_whs = df_whr_2024["Happiness Score"].mean()
mittelwert_whs
```

## Median des Happiness Score (weltweit)

```python
median_whs = df_whr_2024["Happiness Score"].median()
median_whs
```

## Standardabweicheung des Happiness Score (weltweit)

```python
stabw_whs = df_whr_2024["Happiness Score"].std()
stabw_whs
```

## Erstellen eines DFs aus den errechneten Werten

```python
df_whr_stats = pd.DataFrame({
    "Statistik": ["Mittelwert", "Median", "Standardabweichung"],
    "Wert" : [mittelwert_whs, median_whs, stabw_whs]
})

df_whr_stats
```

# Erstellen eines Säulendiagramms zum Vergleich des Happiness Score verschiedener Länder

## Anzeigen des kompletten DF

```python
pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)
df_whr_2024
```

## Filtern der gewünschten Vergleichsländer

```python
auswahl = ["Germany", "Taiwan Province of China", "Republic of Korea", "China", "Hong Kong SAR of China", "Japan"]
df_whr_subset = df_whr_2024[df_whr_2024["Country name"].isin(auswahl)]
df_whr_subset = df_whr_subset.sort_values(by="Happiness Score", ascending=False)
df_whr_subset
```

## Erstellen des Säulendiagramms

```python
plt.figure(figsize=(10, 6))
plt.bar(df_whr_subset["Country name"], df_whr_subset["Happiness Score"], color="skyblue")

plt.title("Vergleich Happiness Score")
plt.xlabel("Land")
plt.ylabel("Happiness Score")
plt.xticks(rotation=0)
plt.tight_layout()

plt.show
```

### Verfeinern des Diagramms (optional)

```python
#Farben definieren
highlighted = "Germany"
farben = ["skyblue" if land != highlighted else "orange" for land in df_whr_subset["Country name"]]
```

### Erstellen des neuen Diagramms

```python
#Säulendiagramm
plt.figure(figsize=(10, 6))
bars = plt.bar(df_whr_subset["Country name"], df_whr_subset["Happiness Score"], color=farben)

#Titel und Achsen
plt.title("Vergleich Happiness Score")
plt.xlabel("Land")
plt.ylabel("Happiness Score")

#Hilfslinien im Hintergrund
plt.grid(axis="y", linestyle="--", alpha=0.5)

#Beschriftung der Balken mit den jeweiligen Werten durch eine Schleife
for bar in bars:
    x = bar.get_x() + bar.get_width() / 2
    y = bar.get_height()
    plt.text(x, y + 0.03, f"{y:.2f}", ha="center")
    
#Achsenlayout
plt.xticks(rotation=0)
plt.tight_layout()

#Anzeigen des Diagramms
plt.show
```
