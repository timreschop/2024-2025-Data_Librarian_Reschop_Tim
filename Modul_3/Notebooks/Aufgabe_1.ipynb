```python
df_sf['Circulation Active Year'] = pd.to_numeric(
    df_sf['Circulation Active Year'], errors='coerce'
)

df_sf['Year Patron Registered'] = pd.to_numeric(
    df_sf['Circulation Active Year'], errors='coerce'
)
```

```python
df_sf['Circulation Active Month'] = pd.to_datetime(
                                        df_sf['Circulation Active Month'],
                                        errors='coerce',
                                        format="%b"
                                    )
```

```python
df_sf['Circulation Active Month'].dt.month
```

```python
df_sf['Membership Duration'] = (df_sf['Circulation Active Year'] - df_sf['Year Patron Registered'])*12 + df_sf['Circulation Active Month'].dt.month
df_sf
```

```python
df_sf['Membership Duration'] = df_sf['Membership Duration'].fillna(0)
df_sf
```
